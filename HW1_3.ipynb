```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
tfinal = int(input()) # Takes final time from user
tinitial = int(input()) # Takes initial time from user
tstep = float(input()) # Takes time difference from user
```

```python
Arraysize = int((tfinal-tinitial)/tstep) # Calculates Total arraysize
Arraysize
```

```python
X = np.zeros(int(Arraysize)) # Array of X initialized
Y = np.zeros(int(Arraysize)) # Array of Y initialized
Theta = np.ones(int(Arraysize)) # Array of Theta initialized
# All the Values are typecasted into integers
X,Y,Theta
```

```python
r = 0.1
L = 1
Wl = np.ones(int(Arraysize)) # Array of Wl initialized
Wr = np.ones(int(Arraysize)) # Array of Wr initialized
```

```python
Wl,Wr
```

```python
for i in range(100):
 Wr[i] = 1
 if i>=5 and i<=15:
   Wr[i] = 12
 if i>=20 and i<=40:
   Wr[i] = 12
 if i>=40 and i<=60:
  Wl[i] = 12
 if i>=60 and i<=80:
  Wl[i] = 12


  # Wr'S & Wl's values are updated according to data from the question.
  # For easier Computation all the values are multiplied by 10. Makes it an integer number.
```

```python
Wl,Wr
```

**state space equations for the differential drive robot model using Eulerâ€™s method**

```python
for i in range(99):
  X[i+1] = X[i] + ((r/2)* (Wl[i]+Wr[i])*(np.cos(Theta[i])))*tstep
  Y[i+1] = Y[i] + ((r/2)* (Wl[i]+Wr[i])*(np.sin(Theta[i])))*tstep
  Theta[i+1] = Theta[i] + ((r/L)* (Wr[i]-Wl[i]))*tstep

    # As the initial values are already given We just need to find the next 99 values.
```

```python
X,Y,Theta
```

```python
plt.plot(X,Y) # Plotting the data

#Add labels and legend
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.title('X VS Y')
```

```python
time = np.linspace(0, 10, len(X))
plt.plot(time , X ) # Plotting the data

#Add labels and legend
plt.xlabel('T')
plt.ylabel('X')
plt.legend()
plt.title('X VS T')
plt.show()
```

```python
time = np.linspace(0, 10, len(Y))
plt.plot(time , Y ) # Plotting the data

#Add labels and legend
plt.xlabel('T')
plt.ylabel('Y')
plt.legend()
plt.title('Y VS T')
plt.show()
```

```python
time = np.linspace(0, 10, len(X))
plt.plot(time,Theta) # # Plotting the data

#Add labels and legend
plt.xlabel('Time values')
plt.ylabel('Theta values')
plt.legend()
plt.title('Theta VS T')
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from matplotlib.animation import PillowWriter
```

```python
#Creating the figure and axes
fig, ax = plt.subplots()
point, = ax.plot([], [], 'b*', markersize=15)

#Add grid lines
ax.grid(True)

def init():
    ax.set_xlim(min(X), max(X))
    ax.set_ylim(min(Y), max(Y))
    return point,

def Revise (AnimationFrame):
    Plot_X = X[AnimationFrame]
    Plot_Y = Y[AnimationFrame]

    point.set_data(Plot_X, Plot_Y )
    return point,

#Set up the animation
animation = FuncAnimation(fig, Revise, frames=len(X), init_func=init, blit=True)

#Save the animation as a GIF
animation.save('custom_animation_with_grid.gif', writer=PillowWriter(fps=10))

from IPython.display import Image
Image(filename='custom_animation_with_grid.gif')
```
